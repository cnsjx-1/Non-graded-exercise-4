```python
import random
import time

card_deck = [
    {"Name": "Diablo", "Health": 100, "Attack": 90, "Defense": 60},
    {"Name": "Medusa", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Jester", "Health": 120, "Attack": 60, "Defense": 90},
    {"Name": "Troll", "Health": 150, "Attack": 40, "Defense": 94},
    {"Name": "Specter", "Health": 100, "Attack": 70, "Defense": 70},
    {"Name": "Mist", "Health": 100, "Attack": 75, "Defense": 65},
    {"Name": "Savage", "Health": 100, "Attack": 90, "Defense": 50},
    {"Name": "Marauder", "Health": 100, "Attack": 85, "Defense": 50},
    {"Name": "Wimp", "Health": 110, "Attack": 40, "Defense": 85},
    {"Name": "Sorcerer", "Health": 100, "Attack": 70, "Defense": 55},
]

deck_size = 5

player_deck = random.sample(card_deck, deck_size)
opponent_deck = random.sample(card_deck, deck_size)

def display_card(card):
    """Displays the card's details."""
    print(f"{card['Name']}: Health = {card['Health']}, Attack = {card['Attack']}, Defense = {card['Defense']}")

def battle_round(attacker, defender):
    """Executes one round of battle between two cards."""
    print(f"\n{attacker['Name']} attacks {defender['Name']}!")
    
    if defender["Defense"] > 0:
        defense_damage = min(defender["Defense"], attacker["Attack"])
        defender["Defense"] -= defense_damage
        remaining_attack = attacker["Attack"] - defense_damage
        print(f"{defender['Name']}'s defense reduced by {defense_damage}. Defense left: {defender['Defense']}")
    else:
        remaining_attack = attacker["Attack"]
    
    if remaining_attack > 0:
        defender["Health"] -= remaining_attack
        if defender["Health"] < 0:
            defender["Health"] = 0
        print(f"{defender['Name']}'s health reduced by {remaining_attack}. Health left: {defender['Health']}")
    
    if defender["Health"] <= 0:
        print(f"{defender['Name']} has been defeated!")
        return False  # Card is defeated and should be removed
    return True  # Card survived the attack

def player_turn():
    """Player's turn to choose a card and attack the opponent."""
    print("\nIt's your turn!")
    print("Your available cards:")
    for i, card in enumerate(player_deck):
        print(f"{i + 1}.", end=" ")
        display_card(card)
    
    choice = int(input(f"\nPick a card to attack (1-{len(player_deck)}): ")) - 1
    player_card = player_deck[choice]
    
    opponent_card = random.choice(opponent_deck)
    print("\nOpponent has chosen:")
    display_card(opponent_card)
    
    if not battle_round(player_card, opponent_card):
        opponent_deck.remove(opponent_card)  # Remove defeated card from opponent's deck
    
def opponent_turn():
    """Opponent's turn to choose a card and attack the player."""
    print("\nIt's the opponent's turn!")
    time.sleep(3)
    
    opponent_card = random.choice(opponent_deck)
    print("Opponent has chosen:")
    display_card(opponent_card)
    
    print("\nYour available cards:")
    for i, card in enumerate(player_deck):
        print(f"{i + 1}.", end=" ")
        display_card(card)
    
    choice = int(input(f"\nPick a card to defend (1-{len(player_deck)}): ")) - 1
    player_card = player_deck[choice]
    
    if not battle_round(opponent_card, player_card):
        player_deck.remove(player_card) 

    # Display opponent's remaining cards and their health values
    print("\nOpponent's remaining cards:")
    for card in opponent_deck:
        display_card(card)

def check_game_over():
    """Check if the game is over based on the decks."""
    if not player_deck:
        print("You have no cards left! You lose!")
        return True
    elif not opponent_deck:
        print("Opponent has no cards left! You win!")
        return True
    return False

print("Welcome to the Card RPG Game!")
while True:
    player_turn()
    if check_game_over():
        break
    opponent_turn()
    if check_game_over():
        break
    print("\nA new round will begin soon...")
    time.sleep(3)

print("Game over!")
```
